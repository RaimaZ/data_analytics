```
#Task 1: Environment set-up
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
%matplotlib inline
import seaborn as sns
#Task 2: Data collection
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv',)
df.head()
```

```
#Task 3: Data wrangling and EDA (exploratory data analysis)
df[(df['county']=='Snohomish')&(df['state']=='Washington')]
df1=df[(df['county']=='Snohomish')&(df['state']=='Washington')]
df1
#Data for the last 30 days for Snohomish county in Washington State
df1.tail(30)
df2=df1.tail(30)
df2
```

```
df2['fips'].isnull()
#no null values to replace; chekced for each column
```

```
#Task 4: Understand NY COVID 19 data in last 30 days
#Show data for NY
df[(df['county']=='Westchester')&(df['state']=='New York')]
df_1=df[(df['county']=='Westchester')&(df['state']=='New York')]
df_1
#Show data for the last 30 days in NY
df_1.tail(30)
df_2=df_1.tail(30)
df_2
```

```
df_2['fips'].isnull()
#no null values to replace; chekced for each column
```

```
sns.jointplot(x='cases',y='deaths',data=df_2, kind='reg')
#Deaths and cases are positively correlated in New York's Westchester County; both have been steadily rising  
```

```
#Show data for the last 30 days for Sheridan county in Wyoming state
df[(df['county']=='Sheridan')&(df['state']=='Wyoming')]
df5=df[(df['county']=='Sheridan')&(df['state']=='Wyoming')]
df5
df5.tail(30)
df5=df5.tail(30)
df5
```

```
#Show data for the last 30 days for Bexar county in Texas state
df[(df['county']=='Bexar')&(df['state']=='Texas')]
df6=df[(df['county']=='Bexar')&(df['state']=='Texas')]
df6
df6.tail(30)
df6=df6.tail(30)
df6
```

```
#Show data for the last 30 days for Cook county in Illinois state
df[(df['county']=='Cook')&(df['state']=='Illinois')]
df7=df[(df['county']=='Cook')&(df['state']=='Illinois')]
df7
df7.tail(30)
df7=df7.tail(30)
df7
```

```
sns.jointplot(x='cases',y='deaths',data=df5, kind='reg')
#Sheridan, Wyoming
```

```
sns.jointplot(x='cases',y='deaths',data=df6, kind='reg')
#Bexar, Texas
```

```
sns.jointplot(x='cases',y='deaths',data=df6, kind='reg')
#Cook, Illinois
```

From looking at these three graphs, we can conclude that the county/state which best handled covid for the last month was Sheridan, Wyoming as cases/death seem to stagnate at certain points. On the other hand, the graphs for Cook, Illinois and Bexar, Texas show a steady increase with no stagnation.
